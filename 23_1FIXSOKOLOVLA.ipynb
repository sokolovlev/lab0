```python
import numpy as np
def func_longest(coords_input:np.ndarray):
    coords = coords_input
    neighbors_point_for_x1 = [coords[1],coords[4]]
    neighbors_point_for_x2 = [coords[2],coords[0]]
    neighbors_point_for_x3 = [coords[3],coords[1]]
    neighbors_point_for_x4 = [coords[4],coords[2]]
    neighbors_point_for_x5 = [coords[0],coords[3]]
    neighbors_points = []
    neighbors_points.append(neighbors_point_for_x1)
    neighbors_points.append(neighbors_point_for_x2)
    neighbors_points.append(neighbors_point_for_x3)
    neighbors_points.append(neighbors_point_for_x4)
    neighbors_points.append(neighbors_point_for_x5)
    sides_info_global = []
    longest_side = []
    i = 0
    k = 0
    for i in range(5):
        points_of_end_of_segment = neighbors_points[i]
        main_point = coords[i]
        for _ in range(2):
            sides_info_local = []
            side = np.sqrt((points_of_end_of_segment[_][0] - main_point[0]) ** 2 + (points_of_end_of_segment[_][1] - main_point[1]) ** 2)
            sides_info_local.append(side)
            sides_info_local.append(i + 1)
            if _ == 0:
                if i < 4:
                    k = i + 2
                else:
                    k = 1
            elif _ == 1:
                if i > 0:
                    k = i
                else:
                    k = i + 5      
            sides_info_local.append(k)
            sides_info_global.append(sides_info_local)
    for _ in range(5):
        side_transmitter = []
        side_transmitter = sides_info_global[_]
        longest_side.append(side_transmitter[0])
    index_of_max_side = longest_side.index(max(longest_side))
    transport_array = sides_info_global[index_of_max_side]
    transport_array.pop(0)
    index_of_point_of_longest_side_end2 = max(transport_array)
    index_of_point_of_longest_side_end1 = min(transport_array)
    coords_of_point_of_longest_side_end2 = coords[index_of_point_of_longest_side_end2 - 1]
    coords_of_point_of_longest_side_end1 = coords[index_of_point_of_longest_side_end1 - 1]
    coords_longest_side = np.array([])
    points_info = []
    points_info.append(index_of_point_of_longest_side_end2)
    points_info.append(index_of_point_of_longest_side_end1)
    coords_longest_side = np.concatenate(([coords_of_point_of_longest_side_end1],[coords_of_point_of_longest_side_end2]))
    coords_longest_side = np.concatenate((coords_longest_side,[points_info]))
    return coords_longest_side



def func_alpha(coords_longest_side_input):
    points_info = coords_longest_side_input[2]
    coords_longest_side = np.delete(coords_longest_side_input, (2), axis = 0)
    delta_y = coords_longest_side[1][1] - coords_longest_side[0][1]
    delta_x = coords_longest_side[1][0] - coords_longest_side[0][0]
    alpha = np.arctan(delta_y / delta_x)   
    transport_list = []
    transport_list.append(alpha)
    transport_list.append(points_info)
    transport_list.append(coords_longest_side)
    return transport_list


def func_transformation(coords:np.ndarray):
    coords1 = coords
    transport_list = func_alpha(func_longest(coords))
    alpha = transport_list[0]
    alpha = alpha - 1.57
    points_info = transport_list[1]
    coords_longest_side = transport_list[2]
    start = coords_longest_side[0]
    end = coords_longest_side[1]
    A = np.array([[np.cos(alpha),-1 * np.sin(alpha)],[np.sin(alpha),np.cos(alpha)]])
    new_start = np.matmul(np.linalg.inv(A),start)
    deleted_index_of_start = 0
    deleted_index_of_end = 0
    for _ in range(5):
        if np.array_equal(coords1[_],start):
            coords1 = np.delete(coords1,(_),axis = 0)
            deleted_index_of_start = _
            break
    for _ in range(4):
        if np.array_equal(coords1[_],end):
            coords1 = np.delete(coords1,(_),axis = 0)
            deleted_index_of_end = _
            break
    new_coords = []   
    for _ in range(3):
        one_point_coords = np.array([coords1[_][0],coords1[_][1]]) 
        one_point_coords = one_point_coords - start
        new_coords.append(one_point_coords)
    new_new_coords = []
    for _ in range(3):
        one_point_coords = np.array([new_coords[_][0],new_coords[_][1]])
        one_point_coords = np.matmul(np.linalg.inv(A),one_point_coords)
        new_new_coords.append(one_point_coords)
    new_coords_inverse = []
    for _ in range(3):
        one_point_inverse = new_new_coords[_]
        one_point_inverse = np.array([-one_point_inverse[0],one_point_inverse[1]])
        new_coords_inverse.append(one_point_inverse)
    coords_return = []
    for _ in range(3):
        one_point_return = new_coords_inverse[_]
        one_point_return = np.matmul(A,one_point_return) + start
        coords_return.append(one_point_return)
    coords_return.insert(deleted_index_of_start,start)
    coords_return.insert(deleted_index_of_end,end)
    result_coord_0 = coords_return[0]
    result_coord_1 = coords_return[1]
    result_coord_2 = coords_return[2]
    result_coord_3 = coords_return[3]
    result_coord_4 = coords_return[4]
    coords_final_return = np.concatenate(([result_coord_0],[result_coord_1],[result_coord_2],[result_coord_3],[result_coord_4]))
    return coords_final_return # выводит двумерный массив где левый столбец это координаты "x",правый это "y"
    
        
    
    
    
    
    
        
        
print(func_transformation([[2,4],[3,2],[4,1],[5,2],[6,4]]))
    
```
